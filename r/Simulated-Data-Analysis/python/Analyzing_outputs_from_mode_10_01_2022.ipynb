### Q5: Visualize the networks: beginning, middle, end. Create different versions of this network structure, where nodes are highlighted by age, race, sex, incarceration status, etc etc. 


```python
agent_time_1 = agent_log_df.loc[agent_log_df['tick'] == 1]
```

```python
# Create graph object
fig, ax = plt.subplots(figsize=(15, 8))

G = nx.from_pandas_edgelist(network_time_1, 'p1', 'p2', create_using=nx.Graph())

# Make types into categories
agent_time_1 = agent_time_1.set_index('agent_id')
agent_time_1 = agent_time_1.reindex(G.nodes())

agent_time_1['agent_race'] = pd.Categorical(agent_time_1['agent_race'])
agent_time_1['agent_race'].cat.codes

# Specify colors
cmap = matplotlib.colors.ListedColormap(['C0', 'darkorange'])

# Draw graph
nx.draw(G, with_labels=True, node_color=agent_time_1['agent_race'].cat.codes, cmap=cmap)
```
